```python
from pathlib import Path

import librosa
import matplotlib.pyplot as plt
```

```python
PATH_DATA = Path().resolve().parent.joinpath('data')
FILE_NAME = 'AUD-20241023-WA0003.mp3'
```

```python
x, sr = librosa.load(PATH_DATA.joinpath('raw').joinpath(FILE_NAME))
```

```python
x
```

```python
sr
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x[:5000], sr=sr)
plt.grid()
```

```python
X = librosa.stft(x)
X_db = librosa.amplitude_to_db(abs(X))
```

```python
plt.figure(figsize=(14, 5))
librosa.display.specshow(X_db, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
plt.figure(figsize=(14, 5))
librosa.display.specshow(X_db, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr)
plt.grid()
```
